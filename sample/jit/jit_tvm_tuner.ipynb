```python
import torch
import torch.nn as nn
from brt.jit.tvm import TVMTuner


class Expert1(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.linear = nn.Linear(512, 1024)

    def forward(self, inputs):
        return self.linear(inputs)


tvm_tuner = TVMTuner()
tvm_tuner.import_pt_netlet(
    "LinearBias",
    "forward",
    Expert1(),
    input_infos={"input0": (1, 512)},
    output_infos={"output0": (1, 1024)},
    parameters={"weight": (512, 1024), "bias": (1, 1024)},
    log_fname="Linear_1_512_1024.json",
)

gird_size, block_size, source_code = tvm_tuner.get_best_template()
```
