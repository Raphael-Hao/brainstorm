```python
import torch

torch.set_printoptions(sci_mode=False)
tensor_all = torch.randint(1,3,(40, 10), dtype=torch.uint8).cuda()
storage_all = tensor_all.storage()._untyped()

tensor_all_32 = torch.tensor([], dtype=torch.float32, device=storage_all.device)
tensor_all_32.set_(storage_all, 0, (10, 10), (10, 1))
print(tensor_all_32)
print(tensor_all_32.data_ptr())


tensor_a = torch.tensor([], dtype=torch.float32, device=storage_all.device)
tensor_b = torch.tensor([], dtype=torch.float32, device=storage_all.device)

tensor_a.set_(storage_all, 0, (5, 10), (10, 1))
tensor_b.set_(storage_all, 5 * 10 , (5, 10), (10, 1))


print(tensor_a)
print(tensor_a.data_ptr())

print(tensor_b)
print(tensor_b.data_ptr())
```

```python
import torch.nn as nn

bn = nn.BatchNorm2d(10)

print(bn.running_mean)
print(bn.running_mean.is_cuda)

running_mean = torch.randn(10, dtype=torch.float32, device="cuda")

bn_running_mean = bn.running_mean
bn_running_mean.data = running_mean

print(bn.running_mean.is_cuda)
```
